```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn.decomposition import PCA
```

```python
x1 = np.arange(-50, 200, 0.5)
x2 = x1 + np.random.normal(scale=20, size=500)
x3 = -3*x1 + np.random.normal(scale=40, size=500)
```

### Plotting the Data

```python
plt.scatter(x=x1, y=x2)
```

```python
plt.scatter(x=x1, y=x3)
```

```python
plt.scatter(x=x2, y=x3)
```

```python
X = np.concatenate([x1.reshape(-1, 1), x2.reshape(-1, 1), x3.reshape(-1, 1)], axis=1)
```

### Performing PCA

```python
pca = PCA(n_components=3)

X_pca = pca.fit_transform(X)
```

```python
plt.scatter(x=X_pca[:, 0], y=X_pca[:, 1])
```

```python
print(f"Variance of the columns in X:     {X[:, 0].var(): 6.2f} {X[:, 1].var(): 6.2f} {X[:, 2].var(): 6.2f}")
print(f"Variance of the columns in X PCA: {X_pca[:, 0].var(): 6.2f} {X_pca[:, 1].var(): 6.2f} {X_pca[:, 2].var(): 6.2f}")
```

```python
X_2 = np.concatenate([x1.reshape(-1, 1), x2.reshape(-1, 1)], axis=1)
```

```python
pca = PCA(n_components=2)

X_pca_2 = pca.fit_transform(X_2)
```
