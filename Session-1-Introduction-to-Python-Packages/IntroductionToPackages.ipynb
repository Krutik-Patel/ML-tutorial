##### Dimensions and Shaping

As numpy is used to make vectors, matrices and tensors, dimensions become very important

NOTE: (n, ) and (n, 1) are not the same shapes!


```python
arr_1 = [1, 2, 3, 4, 5, 6, 7, 8]
arr_2 = [
    [1, 2, 3, 4],
    [5, 6, 7, 8],
]
arr_3 = [
    [
        [1, 2],
        [3, 4]
    ],
    [
        [5, 6],
        [7, 8]
    ]
]

np_arr_1 = np.array(arr_1)
np_arr_2 = np.array(arr_2)
np_arr_3 = np.array(arr_3)

print(np_arr_1.shape)
print(np_arr_2.shape)
print(np_arr_3.shape)
```

np.zeros, np.ones, np.full, np.empty have their analogues np.zeros_like, np.ones_like, np.full_like, np.empty_like to make arrays which take the shape of some other given array. The datatype is also taken by the new array unless specified


```python
np.zeros_like(np_arr_3).shape
```

##### Reshaping

In ML, especially Deep Learning, we need to reshape arrays to fit our model.

The new shape must be compatible with the older shape i.e. the number of elements should match with the new shape


```python
print("Original Array:\n", np_arr_3)
print()
print("2D Array:\n", np_arr_3.reshape(2, 4))
print()
print("1D Array:\n", np_arr_3.reshape(8,))
```

When we don't care about one of the dimensions, we can keep it unspecified for numpy to adjust accordingly. For example, if the original shape is (8, ) and we need to keep the first dimension as 2, while not caring about the second dimension, we can use (2, -1) to specify the new shape.


```python
print("Original Array:\n", np_arr_3)
print()
print("2D Array:\n", np_arr_3.reshape(2, 4))
print()
print("2D Array without specifyng the second dimension:\n", np_arr_3.reshape(2, -1))
print()
```

## Broadcasting

Numpy's broadcasting feature allows you to perform arithmetic operations on arrays of different shapes. It essentially "stretches" the smaller array across the larger array so that they have compatible shapes. This makes vectorized operations more powerful and can lead to significant speed improvements.

### Rules of Broadcasting

1. Start with the trailing (i.e., rightmost) dimensions of the two arrays and compare them element-wise.
2. Two dimensions are compatible when:
   - They are equal, or
   - One of them is 1.
3. If the dimensions are not compatible and one of them is not 1, an error is raised.
4. If one of the dimensions is 1, it can be stretched to match the other dimension.


### Examples using Broadcasting


```python
arr = np.array([1, 2, 3])
scalar = 2

print("Array:", arr)
print("Scalar:", scalar)
print("Broadcasted result:", arr + scalar) 
```

```python
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
arr_1d = np.array([1, 2, 3])

print("2D Array:\n", arr_2d)
print("1D Array:", arr_1d)
print("Broadcasted result:\n", arr_2d + arr_1d) 
```

```python
arr1 = np.ones((3, 1, 4))
arr2 = np.random.random_sample((2, 1))

print("1st Array:\n", arr1)
print("2nd Array:", arr2)
print("Broadcasted result:\n", arr1 + arr2)
print("Broadcasted result shape:\n", (arr1 + arr2).shape)
```

## `apply` Method


The apply method in Pandas is a powerful tool that allows you to apply a function along an axis of the DataFrame. This can be used to perform complex operations row-wise or column-wise in a very flexible manner.


```python
def to_thousands(price):
    return price / 1000

housing['price_thousands'] = housing['price'].apply(to_thousands)

housing.head()
```

```python
housing['size_category'] = housing['bedrooms'].apply(lambda x: 'large' if x > 3 else 'small')

housing.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Basic Plotting


### Line Plot

A line plot is one of the simplest and most common types of plots. It is used to display data points along a line.


```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### Scatter Plot

A scatter plot is used to show the relationship between two variables by plotting them as points in 2D space.


```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### Bar Plot

A bar plot is used to display categorical data with rectangular bars representing the quantity of each category.


```python
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

plt.bar(categories, values)
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

### Histogram

A histogram is used to display the distribution of a dataset. It divides the data into bins and counts the number of data points that fall into each bin.


```python
data = np.random.randn(1000)

plt.hist(data, bins=30, edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## Cutomizing Plots


### Adding Labels and Titles


```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='Sine')
plt.title('Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

### Changing Line Styles and Colors


```python
plt.plot(x, y, color='red', linestyle='--', marker='o')
plt.title('Styled Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### Subplots


```python
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title('Sine Wave')

plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x))
plt.title('Cosine Wave')

plt.tight_layout()
plt.show()
```
